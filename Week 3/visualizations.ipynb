```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px

# Load the car.csv dataset
cars = pd.read_csv("cars.csv")
```

```python
# View the first few rows of the dataset
print(cars.head())

# Get summary statistics
print(cars.describe())

# Check data types
print(cars.dtypes)
```

```python
# Line plot using Matplotlib
weight = cars['MPG']
weight.plot()
```

```python
# Scatter plot using Seaborn
sns.scatterplot(data=cars, x='Weight', y='Acceleration' )
plt.title('Weight vs. Acceleration')
plt.show()
```

```python
# Histogram using Plotly
fig = px.histogram(cars, x='Horsepower')
fig.update_layout(title='Horse Power', xaxis_title='HorsePower', yaxis_title='Count')
fig.show()
```

```python
# Bar plot using Matplotlib
plt.bar(cars['Car'], cars['Origin'])
plt.xlabel('Car')
plt.ylabel('Origin')
plt.title('Car by Origin')
plt.xticks(rotation=90)
plt.show()
```

```python
cars['Weight'].plot(x=cars['Car'].value_counts,
        kind='bar',
        stacked=True,
        title='Car Count')
```

```python
category_count = cars['Car'].value_counts(cars['Origin'].value_counts, sort=True)
plt.rcParams['figure.figsize'] = [6, 12]
category_count.plot.barh()
```

```python
# Box plot using Seaborn
sns.boxplot(data=cars, x='Origin', y='Weight')
plt.title('Cars Distribution by Weight')
plt.show()
```

```python
# Box plot using Seaborn
sns.boxplot(data=cars, x='Origin', y='Horsepower')
plt.title('Car Distribution by HorsePower')
plt.show()
```

```python
# Box plot using Seaborn
sns.boxplot(data=cars, x='Origin', y='Acceleration')
plt.title('Cars Distribution by Acceleration')
plt.show()
```
